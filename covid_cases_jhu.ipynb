```python
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib
from scipy.stats import gamma
import numpy as np
import c3aidatalake
```

```python
rural_counties = pd.read_csv('final_rural_counties.csv')
rural_counties
```

```python
today = pd.Timestamp.now().strftime("%Y-%m-%d")

monthly_rural_cases = c3aidatalake.evalmetrics(
    "outbreaklocation",
    {
        "spec" : {
            "ids" : list(rural_counties['c3_data_id']),
            "expressions" : ["JHU_ConfirmedCases"],
            "start" : "2020-02-24",
            "end" : today,
            "interval" : "MONTH",
        }
    },
    get_all = True
)

monthly_rural_cases
```

```python
monthly_rural_cases
```

```python
months = monthly_rural_cases.dates.dt.month_name()
monthly_rural_cases['month'] = months
```

```python
rural_cases = monthly_rural_cases.set_index('month').T
rural_cases = rural_cases.reset_index()
monthly_r_cases = rural_cases[rural_cases['index'].str.endswith(".data")]
```

```python
monthly_r_cases 
```

```python
#getting parent_id from id col
id_names = monthly_r_cases['index'].tolist()
split_names = [str(id_names[x]).replace(".JHU_ConfirmedCases.data", "")for x in range(len(id_names))]
monthly_r_cases['index'] = split_names
monthly_r_cases 
```

```python
rural_mcases = rural_counties.merge(monthly_r_cases, left_on='c3_data_id', right_on = "index").drop(columns = ['Unnamed: 0', 'county', 'state', 'c3_data_id', 'index'])
rural_mcases
```
