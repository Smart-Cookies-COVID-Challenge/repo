```python
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib
from scipy.stats import gamma
import numpy as np
import c3aidatalake
```

```python
rural_counties = pd.read_csv('final_rural_counties.csv')
rural_counties
```

```python
today = pd.Timestamp.now().strftime("%Y-%m-%d")

monthly_rural_cases = c3aidatalake.evalmetrics(
    "outbreaklocation",
    {
        "spec" : {
            "ids" : list(rural_counties['c3_data_id']),
            "expressions" : ["JHU_ConfirmedCases"],
            "start" : "2020-02-24",
            "end" : today,
            "interval" : "MONTH",
        }
    },
    get_all = True
)

monthly_rural_cases
```

```python
monthly_rural_cases
```

```python
rural_cases = monthly_rural_cases.set_index('dates').T
rural_cases = rural_cases.reset_index()
monthly_r_cases = rural_cases[rural_cases['index'].str.endswith(".data")]
```

```python
monthly_r_cases.shape
```

```python
key = [monthly_r_cases.columns[i] for i in range(11)]
value = ['id', 'feb_cases', 'mar_cases', 'apr_cases','may_cases','june_cases', 'july_cases','aug_cases',
         'sept_cases', 'oct_cases', 'nov_cases']
col_dict = {k:v for (k,v) in zip(key,value)}
```

```python
rural_monthly = monthly_r_cases.rename(columns = col_dict)
rural_monthly
```

```python
#getting parent_id from id col
id_names = rural_monthly['id'].tolist()
split_names = [str(id_names[x]).replace(".JHU_ConfirmedCases.data", "")for x in range(len(id_names))]

rural_monthly['id'] = split_names
rural_monthly
```

```python
rural_mcases = rural_counties.merge(rural_monthly, left_on='c3_data_id', right_on = "id")
rural_mcases
```

## Is there Data available?

```python
today = pd.Timestamp.now().strftime("%Y-%m-%d")

jhucases = c3aidatalake.evalmetrics(
    "outbreaklocation",
    {
        "spec" : {
            "ids" : ['St.Clair_Missouri_UnitedStates', 'Ste.Genevieve_Missouri_UnitedStates'],
            "expressions" : ["JHU_ConfirmedCases"],
            "start" : "2020-02-24",
            "end" : today,
            "interval" : "MONTH",
        }
    },
    get_all = True
)

jhucases
```

```python
missingdf = jhucases.set_index('dates').T
missing_df = missingdf.reset_index()
missing_df = missing_df[missing_df['index'].str.endswith(".data")]
key = [missing_df.columns[i] for i in range(11)]
value = ['id', 'feb_cases', 'mar_cases', 'apr_cases','may_cases','june_cases', 'july_cases','aug_cases',
         'sept_cases', 'oct_cases', 'nov_cases']
col_dict = {k:v for (k,v) in zip(key,value)}
missing_monthly = missing_df.rename(columns = col_dict)
missing_monthly
```

#### Was cut because there is an extra period in the name of the County

```python
missing_id_names = missing_monthly['id'].tolist()
missing_split_names = [str(missing_id_names[x]).replace(".JHU_ConfirmedCases.data", "")for x in range(len(missing_id_names))]
missing_split_names
missing_monthly['id'] = missing_split_names
missing_monthly
```

```python
rural_missing = rural_counties[rural_counties['c3_data_id'].isin(missing_split_names)]
rural_missing = rural_missing.merge(missing_monthly, left_on = 'c3_data_id', right_on = 'id')
```

## Appending to Original Table

```python
full_rural = pd.concat([rural_mcases, rural_missing]).sort_values(by = 'id')
full_rural.reset_index(drop = True)
full_rural_cases = full_rural.drop(columns = ['Unnamed: 0', 'id'])
full_rural_cases
```

## Saving Cumulative Monthly Cases to CSV Format

```python
full_rural_cases.to_csv('monthly_rural_cumulative_cases.csv', index = False)
```

## New Monthly Cases

```python
df = full_rural_cases.copy()
df
```

```python
month_col = ['feb_cases', 'mar_cases',
       'apr_cases', 'may_cases', 'june_cases', 'july_cases', 'aug_cases',
       'sept_cases', 'oct_cases', 'nov_cases']
new_col = ['feb_new', 'mar_new',
       'apr_new', 'may_new', 'june_new', 'july_new', 'aug_new',
       'sept_new', 'oct_new', 'nov_new']

df['feb_new'] = df['feb_cases']
for i in range(len(month_col)-1):
    df[new_col[i+1]] = df[month_col[i+1]] - df[month_col[i]]

df = df.drop(columns = month_col)
df
```

## Saving New Cases to CSV Format

```python
df.to_csv('monthly_rural_new_cases.csv', index = False)
```
